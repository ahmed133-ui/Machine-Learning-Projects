# **Introduction**
This model was developed to address a nutritional challenge faced by a petting zoo, which was unable to accurately predict weight changes in penguins based on species, gender, and individual characteristics. By analyzing the correlation between body mass and factors such as bill length, gender, and species, the model provides precise predictions of body mass, enabling the zoo to tailor nutritional plans to the unique needs of each penguin. This approach not only enhances care but also improves the health and well-being of the animals.

```python
## Load the Libraries

import pandas as pd
import seaborn as sns

## Load the Dataset
penguins = sns.load_dataset("penguins")

## Perform EDA
penguins=penguins[["body_mass_g","bill_length_mm","sex","species"]]
penguins.columns=["body_mass_g","bill_length_mm","gender","species"]
penguins.dropna(inplace=True)
penguins.reset_index(drop=True,inplace=True)
penguins.head()
```

```python
## Assign the X & Y variables
penguins_x=penguins[["bill_length_mm","gender","species"]]
penguins_y=penguins[["body_mass_g"]]

## Load & Train the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(penguins_x,penguins_y,test_size=0.3,random_state=42)
ols_formula="body_mass_g~bill_length_mm+C(gender)+C(species)"
from statsmodels.formula.api import ols
ols_data=pd.concat([x_train,y_train],axis=1)
OLS=ols(ols_formula,data=ols_data)
model=OLS.fit()
model.summary()
```
